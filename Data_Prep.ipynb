```python
df = pd.read_csv('video_dataset.csv')
```

```python
columns_to_remove = [
    "contract_address", "chain_id", "chart_url", "user_type",
    "slug", "identifier", "first_name", "last_name", "username",
    "is_locked", "has_evm_wallet", "has_solana_wallet", "following",
    "source_matrix", "gif_thumbnail_url"
]

# Drop the columns
df_cleaned = df.drop(columns=columns_to_remove, axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder

df = df_cleaned.dropna(subset=["baseToken"])

# Encode user and video IDs
user_encoder = LabelEncoder()
video_encoder = LabelEncoder()

df["user_id"] = user_encoder.fit_transform(df["baseToken"])
df["video_id"] = video_encoder.fit_transform(df["id"])
df.drop(columns=["baseToken", "id"], inplace=True)
```

```python
df["interaction"] = df["upvoted"].astype(int) + df["bookmarked"].astype(int)
```

```python
df = df[df["interaction"] > 0]
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df["interaction"] = scaler.fit_transform(df[["interaction"]])
```

```python
# Convert boolean values to integers
df["upvoted"] = df["upvoted"].astype(int)
df["bookmarked"] = df["bookmarked"].astype(int)
df["interaction"] = df["upvoted"] + df["bookmarked"]
df = df[df["interaction"] > 0]
df = df[["user_id", "video_id", "interaction"]]
```

```python
from sklearn.model_selection import train_test_split

train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import LabelEncoder

# Re-encode user and video IDs
user_encoder = LabelEncoder()
video_encoder = LabelEncoder()

df["user_id"] = user_encoder.fit_transform(df["user_id"])
df["video_id"] = video_encoder.fit_transform(df["video_id"])

# Verify the encoding
print("Unique users:", df["user_id"].nunique())
print("Unique videos:", df["video_id"].nunique())
print("Max video_id:", df["video_id"].max())  # Should now be 25 (0 to 25)
```

```python
from transformers import pipeline

sentiment_pipeline = pipeline("sentiment-analysis")

def analyze_sentiment(text):
    result = sentiment_pipeline(text)[0]
    label = result.get('label', 'NEUTRAL')  
    sentiment_scores = {
        'POSITIVE': 1,
        'NEGATIVE': -1,
        'NEUTRAL': 0  
    }
    return sentiment_scores.get(label.upper(), 0)

df["mood_label"] = df_cleaned["title"].fillna("").apply(analyze_sentiment)
```
